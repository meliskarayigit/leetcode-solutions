```python
# What are the average earnings broken down by COUNTRY, SOURCE, GENDER, and AGE?

earning= df.groupby(['COUNTRY', 'SOURCE', 'SEX','AGE'])['PRICE'].mean()
agg_df= earning.sort_values(ascending=False)
agg_df= agg_df.reset_index()
print(agg_df)

```

```python
agg_df['age_cat']=pd.cut(agg_df['AGE'], (0,18,23,30,40,70))
agg_df
```

```python
#level-based customers (personas)

cols_to_combine = ['COUNTRY', 'SOURCE', 'SEX', 'age_cat']
separator = '_'

agg_df['customers_level_based']= agg_df[cols_to_combine].astype(str).agg('_'.join,axis=1)
agg_df.head()
customers_level_based= agg_df.groupby('customers_level_based')['PRICE'].mean()
customers_level_based
```

```python
agg_df['SEGMENT']= pd.qcut(agg_df['PRICE'], 4,labels=['D','C','B','A'])
agg_list=['mean','max','min','sum']
agg_df.groupby('SEGMENT')['PRICE'].agg(agg_list)
```
